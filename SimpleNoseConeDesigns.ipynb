# Simple Nose Cone Models:
This file depicts 2D models of Conical, Ogive, and Elliptical nose cone shapes.

These nose cone shapes are typically found in model and (lower-powered) High-Power Rockets.

```python
import numpy as np
import math
import NoseConeEquations as nosecone
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#USER SHOULD CHANGE VALUES:
unit = 'cm'
nc_radius = 8
nc_length = 60
```

## Conical:
Popular due to ease of manufacture and drag characteristics

```python
#DONT MODIFY
#Creating Conical Nose Cone 2D Model
x = np.linspace(0, nc_length, num=11)
y_top = nosecone.conical_shape(x, nc_radius, nc_length)
y_bot = nosecone.conical_shape(x*-1, nc_radius, nc_length)
cl = x * 0
plt.plot(x, y_top,'k')
plt.plot(x, y_bot, 'k')
plt.plot(x, cl, 'k-.')
plt.grid()
plt.title("Conical Nose Cone Model")
plt.xlabel("Length ({})".format(unit))
plt.ylabel("Radius ({})".format(unit))
```

## Tangent Ogive:
Popular due to ease of construction

```python
#DONT MODIFY
#Creating 2D Model of Ogive Nose Cone
x = np.linspace(0, nc_length, num=11)
y_top = [nosecone.tangent_ogive_shape(x, nc_radius, nc_length) for x in x]
y_bot = [-1*nosecone.tangent_ogive_shape(x, nc_radius, nc_length) for x in x]
plt.plot(x, y_top,'k')
plt.plot(x, y_bot, 'k')
plt.plot(x, cl, 'k-.')
plt.grid()
plt.title("Ogive Nose Cone Model")
plt.xlabel("Length ({})".format(unit))
plt.ylabel("Radius ({})".format(unit))
plt.show()
```

## Elliptical:
Popular in subsonic flight due to blunt nose and tangent base.

Not normally found in professional rocketry.

```python
#DONT MODIFY
#Creating 2D Model of Elliptical Nose Cone
x = np.linspace(0, nc_length, num=101)
y_top = [nosecone.elliptical_shape(x, nc_radius, nc_length) for x in x]
y_bot = [-1*nosecone.elliptical_shape(x, nc_radius, nc_length) for x in x]
cl = x * 0
plt.plot(x, y_top, 'k')
plt.plot(x, y_bot, 'k')
plt.plot(x, cl, 'k-.')
plt.grid()
plt.title("Elliptical Nose Cone Model")
plt.xlabel("Length ({})".format(unit))
plt.ylabel("Radius ({})".format(unit))
plt.show()
```
